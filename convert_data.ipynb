```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("raw_data.csv")
```

```python
data['Period']=pd.to_datetime(data.Period)
```

```python
new_data = pd.DataFrame(pd.date_range(data.Period[0],data.Period[98],freq='d'),columns = ['Period'])
```

```python
def convert_df(df, column):
    ret_lis = []
    for i in range(len(df)-1):
        ret_lis = ret_lis + list(np.linspace(df[column][i], df[column][i+1],
                                             num=df[["Period"]].Period[i].days_in_month+1))[:-1]
    return ret_lis + [df[column].values[-1]]
```

```python
for col in data.columns.drop('Period'):
    new_data[col] = convert_df(data,col)
```
